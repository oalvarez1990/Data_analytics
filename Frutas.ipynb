# Definir la lista de listas

```python
frutas = [
    ["Manzana", "Roja", "Dulce"],
    ["Plátano", "Amarillo", "Dulce"],
    ["Lima", "Verde", "Ácida"]
]
```

# Escribir en un archivo CSV

```python
import csv
with open('./content/frutas.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(frutas)

print("Archivo CSV creado y guardado en ./content/frutas.csv")
```

```python
import pandas as pd
df = pd.read_csv('./content/frutas.csv', header=None)
print(df)
```

# Contar la cantidad de frutas por color y sabor

```python
from collections import Counter

# Contar la cantidad de frutas por color
colores = [fruta[1] for fruta in frutas]
contador_colores = Counter(colores)

# Contar la cantidad de frutas por sabor
sabores = [fruta[2] for fruta in frutas]
contador_sabores = Counter(sabores)

print("Contador de colores:", contador_colores)
print("Contador de sabores:", contador_sabores)
```

# Crear gráfico de barras para colores

```python
import matplotlib.pyplot as plt

# Crear gráfico de barras para colores
plt.figure(figsize=(10, 5))

# Colores
plt.subplot(1, 2, 1)
plt.bar(contador_colores.keys(), contador_colores.values(),
        color=['red', 'yellow', 'green'])
plt.xlabel('Color')
plt.ylabel('Cantidad')
plt.title('Cantidad de Frutas por Color')

# Sabores
plt.subplot(1, 2, 2)
plt.bar(contador_sabores.keys(), contador_sabores.values(),
        color=['blue', 'orange'])
plt.xlabel('Sabor')
plt.ylabel('Cantidad')
plt.title('Cantidad de Frutas por Sabor')

plt.tight_layout()
plt.show()
```
