```python
import numpy as np
import matplotlib.pyplot as plt

class PolinomioGrapher:
    def __init__(self):
        self.polinomios = []

    def agregar_polinomio(self, a, n):
        """
        Agrega un polinomio a la lista de polinomios a graficar.
        
        Args:
            a (float): Coeficiente del polinomio.
            n (int): Exponente del polinomio.
        """
        self.polinomios.append((a, n))

    def graficar(self):
        """
        Grafica todos los polinomios almacenados en la lista.
        """
        x = np.linspace(-10, 10, 400)
        plt.figure(figsize=(10, 6))
        
        for polinomio in self.polinomios:
            a, n = polinomio
            y = a * x**n
            plt.plot(x, y, label=f'{a} * x^{n}')
        
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Gr√°fica de Polinomios')
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
# Crear una instancia de PolinomioGrapher
grapher = PolinomioGrapher()

# Agregar polinomios
grapher.agregar_polinomio(1, 2)
grapher.agregar_polinomio(2, 3)
grapher.agregar_polinomio(-1, 1)
grapher.agregar_polinomio(0.5, 4)

# Graficar los polinomios
grapher.graficar()
```
